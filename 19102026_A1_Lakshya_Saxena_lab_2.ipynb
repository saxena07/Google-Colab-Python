###2.2 Plotting a line

#### Solution

###2.3 Plotting two or more lines on same plot
####Problem: Plot two lines on same plot of different colours

####Solution

```python
import matplotlib.pyplot as plt
  
# line 1 points
x1 = [1,2,3,4,5]
y1 = [2,4,1,3,5]
# plotting the line 1 points 
plt.plot(x1, y1, label = "line 1")
  
# line 2 points
x2 = [1,2,3,4,5]
y2 = [4,1,3,2,5]
# plotting the line 2 points 
plt.plot(x2, y2, label = "line 2")
  
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
# giving a title to my graph
plt.title('Two lines on same plot')
  
# show a legend on the plot
plt.legend()
  
# function to show the plot
plt.show()
```

#### Note
How to differentiate between the lines: By using a name(label) which is passed as an argument of *.plot()* function.
The small rectangular box giving information about type of line and its color is called legend. 
* Add a legend to plot using *.legend()* function.

### 2.4 Plotting curves 
#### Problem
Plotting curve for sin(x)

#### Solution


```python
# importing the required modules
import matplotlib.pyplot as plt
import numpy as np
  
# setting the x - coordinates
x = np.arange(0, 20*(np.pi), 0.1)
# setting the corresponding y - coordinates
y = np.sin(x)
  
# potting the points
plt.plot(x, y, label ='sin(x)')

# naming the x axis
plt.xlabel('x - axis')

# naming the y axis
plt.ylabel('y(x)=sin(x)')

# giving a title to my graph
plt.title('Sine curve')

# show a legend on the plot
plt.legend()

# function to show the plot
plt.show()




```

#### Note
Here, NumPy used (a general-purpose array-processing package in python).

To set the x – axis values, use *np.arange()* method used in which first two arguments are for range and third one for step-wise increment like *linspace* from MATLAB. The result is a numpy array.
To get corresponding y-axis values, predefined np.sin() used on the numpy array.
Finally, the points are plotted by passing x and y arrays to the *plt.plot()* function.

```python
# view sine array values with size
s=np.sin(x)
print(s)
print('\n Size of sine array is',s.shape)
```

###2.5 Plotting with keyword strings
####Problem: To plot particular variables with strings

####Solution
For example, with numpy.recarray or pandas.DataFrame.

Matplotlib allows you provide such an object with the data keyword argument. If provided, then you may generate plots with the strings corresponding to these variables.

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

#scatter plot
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

###2.6 Stem plot for Discrete-time Signals
####Problem: To generate a discrete time sinusoidal signal.

###Solution
matplotlib.pyplot.stem() creates stem plots. A Stem plot plots vertical lines at each x position covered under the graph from the baseline to y, and places a marker there.



```python
### importing libraries
import matplotlib.pyplot as plt


import numpy as np

## setting the x - coordinates
n = np.linspace(0, 30, 31)

# calculating corresponding the y - coordinates
xn = (np.sin(2*n*(np.pi)/15))

# plotting the points
plt.stem(n, xn, use_line_collection = True)
# naming the x axis
plt.xlabel('n (samples)')
# naming the y axis
plt.ylabel('x[n]')
# giving a title to my graph
plt.title('DT Sine curve')

#showing the plot
plt.show()
```

####Problem: To generate a discrete time unit impulse and unit step signal.
Solution

```python
### importing libraries
import matplotlib.pyplot as plt
import numpy as np
  
n = np.linspace(-5, 5, 11)
delta = 1*(n==0)
u = 1*(n>=0)
  
plt.stem(n, delta, use_line_collection = True)
# naming the x axis
plt.xlabel('n')
# naming the y axis
plt.ylabel('x[n] = delta[n]')

# giving a title to my graph
plt.title('Unit Impulse Sequence')
plt.show()

plt.stem(n, u, use_line_collection = True)
# naming the x axis
plt.xlabel('n')
# naming the y axis
plt.ylabel('x[n] = u[n]')

# giving a title to my graph
plt.title('Unit Step Sequence')
plt.show()
```

### 2.7.1 Subplots and Plotting with categorical variables
#### Problem: create a plot using categorical variables. 

#### Solution: Matplotlib allows you to pass categorical variables directly to many plotting functions.

### 2.7.2 Working with multiple figures and axes
MATLAB, and pyplot, have the concept of the current figure and the current axes. All plotting functions apply to the current axes. The function gca returns the current axes (a matplotlib.axes.Axes instance), and gcf returns the current figure (a matplotlib.figure.Figure instance). 

####Problem: To create two subplots.
####Solution

```python
#creating function for generating signal using diffrent time ranges
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.1)

plt.figure()
#subplot matrix size 2 x 1, window 1

plt.subplot(211)
plt.plot(t1, f(t1), 'gx', t2, f(t2), 'y-')

#subplot matrix size 2 x 1, window 2
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'g-')

plt.show()
```
